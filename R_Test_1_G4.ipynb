```r
pdata <- read.table("/content/heart (2).csv", 
                  header=T, 
                  sep=",",
                  strip.white=T,
                  stringsAsFactors = T)

pdata
```

```r
summary(pdata)
```

```r
library(ggplot2)
```

```r
library(dplyr)
```

```r
pdatasexM <- filter(pdata, Sex=="M")
pdatasexMH <- arrange(pdatasexM, HeartDisease)
pdatasexMH
```

```r
pdatasexF <- filter(pdata, Sex=="F")
pdatasexFH <- arrange(pdata, HeartDisease)
pdatasexFH
```

```r
pdataheartF <- filter(pdatasexF, HeartDisease=="1")
pdataheartF
```

```r
pdataheartM <- filter(pdatasexM, HeartDisease=="1")
pdataheartM
```

```r
pdataAge <- arrange(pdata,Age)

heartsummary <- select(pdataAge, Age, Sex, Cholesterol, RestingBP, RestingECG, Oldpeak, ST_Slope, HeartDisease)
heartsummary
```

```r
#pie chart between sex and heart disease

ggplot(pdata, aes(x="", y=HeartDisease, fill=Sex)) +
  geom_bar(stat="identity", width=1) +
  coord_polar("y", start=0)
```

```r
#pie chart Heart Disease and the type of Chest Pain Type

ggplot(pdata, mapping = aes(x=HeartDisease,fill =ChestPainType)) +
  geom_bar(position = "dodge") +
  labs(title = "Heart Disease and the type of Chest Pain Type", x ="Heart Disease", y = "Count", fill="CHest Pain Type")
```

```r
#line chart for Correlation of Chest Pain Type, Age and Heart Disease

pdatasexFH %>%
  ggplot( aes(x=HeartDisease, y=Age, group=ChestPainType, color=ChestPainType)) +
    geom_line()+
    labs(title = "Correlation of Chest Pain Type, Age and Heart Disease")
```

```r
#Female Patients have Heart Disease by age

ggplot(pdataheartF, aes(x = Age)) +
  geom_histogram(fill = "cornflowerblue", 
                 color = "white") + 
  labs(title="Female Patients have Heart Disease by age",
       x = "Age")

#Male Patients have Heart Disease by age

ggplot(pdataheartM, aes(x = Age)) +
  geom_histogram(fill = "cornflowerblue", 
                 color = "white") + 
  labs(title="Male Patients have Heart Disease by age",
       x = "Age")
```

```r
#point graph based Cholesterol, Sex and Age

ggplot(data = pdata, 
       aes(x = Age, y = Cholesterol, col = Sex)) + geom_point()+
       labs(title = "Cholesterol, Sex and Age")
```

```r
ggplot(pdata, aes(x =as.factor(Age), fill= as.factor(HeartDisease) )) + 
  geom_bar( ) +
  scale_fill_brewer(palette = "Set1") +
  theme(legend.position="none")+
  labs(title = "Heart Disease and Sex", x ="Age", y = "Count", fill="HeartDisease")
```

```r
ggplot(pdatasexFH, aes(x=as.factor(Age), fill=as.factor(HeartDisease) )) +  
  geom_bar( ) +
  scale_fill_manual(values = c("red", "green", "blue") ) +
  theme(legend.position="none")
```
